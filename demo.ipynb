````python
AUDIO_URL = "https://keithito.com/LJ-Speech-Dataset/LJ025-0076.wav"
````

````python
from huggingface_hub import hf_hub_download
checkpoint_path = hf_hub_download(
    repo_id="Finikita255/SoundStream-checkpoints",
    filename="checkpoints/checkpoint_final.pth",
)
print(f"чекпоинт скачан: {checkpoint_path}")
````

````python
import sys
sys.path.insert(0, '/content/DLHW4')

import torch
import torchaudio
import urllib.request
from IPython.display import Audio, display

from inference import load_model_from_checkpoint, load_audio_from_path, run_trough_codec
````

Скачиваем аудио по ссылке

````python
urllib.request.urlretrieve(AUDIO_URL, "input.wav")
````

Загружаем нашу модель с HF и прогоняем кодек

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder, decoder, rvq = load_model_from_checkpoint(checkpoint_path, device)
waveform = load_audio_from_path("input.wav", device)
reconstructed = run_trough_codec(waveform, encoder, decoder, rvq)
print("Оригинал:")
display(Audio(waveform[0, 0].cpu().numpy(), rate=16000))
print("После кодека:")
display(Audio(reconstructed[0, 0].cpu().numpy(), rate=16000))
````
